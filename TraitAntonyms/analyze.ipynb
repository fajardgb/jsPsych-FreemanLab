```python
def get_trait(str):
    #find words betwen <strong>word</strong>
    matches = re.findall(r'<strong>(.*?)</strong>', str, re.IGNORECASE)
    if matches:  
        return matches[1] #get 2nd string - this is the question!
    else:
        return ''


def get_response(str):
    #find words between ""
    matches = re.findall(r'"(.*?)"', str, re.IGNORECASE)
    if matches:
        if matches[0] == 'Q0':
            return matches[1]
        else: 
            return str
    else: 
        return ''
```
